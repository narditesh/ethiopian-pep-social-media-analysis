```python
import pandas as pd

df = pd.read_csv("Ethiopian PEP Social Media Verification Dataset  - Dataset.csv", skiprows=1)
df.head()
```

```python
df.columns
```

```python
df = df.dropna(how='all')  # remove empty rows
df = df.reset_index(drop=True)
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
```

```python
df['Verified X (Twitter) Account (Yes/No)'].value_counts()
```

```python
df['X (Twitter) Link'].notnull().sum()
df['Facebook Link'].notnull().sum()
```

```python
df['Confidence'].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
df['Verified X (Twitter) Account (Yes/No)'].value_counts().plot(kind='bar')
plt.title("Verified vs Not Verified X (Twitter) Accounts")
plt.xlabel("Verification Status")
plt.ylabel("Count")
plt.show()
```

```python
df['Confidence'].value_counts().plot(kind='bar')
plt.title("Confidence Levels Distribution")
plt.xlabel("Confidence Level")
plt.ylabel("Count")
plt.show()
```

```python
df[(df['X (Twitter) Link'] == "Unknown") & (df['Facebook Link'] == "Unknown")]
```

```python
df[df['Verified X (Twitter) Account (Yes/No)'] == "Yes"]
```

```python
x_count = df['X (Twitter) Link'].apply(lambda x: x != "Unknown").sum()
fb_count = df['Facebook Link'].apply(lambda x: x != "Unknown").sum()

print("X accounts:", x_count)
print("Facebook accounts:", fb_count)
```

```python
df.to_csv("cleaned_dataset.csv", index=False)
```
